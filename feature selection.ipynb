```python
#information gain
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\orange pc\Desktop\weather.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder

df['outlook'].unique()

df['humidity'].unique()
df['windy'].unique()
```

```python
lbl_encode=LabelEncoder()
```

```python
outlooks=lbl_encode.fit_transform(df['outlook'])
outlooks
```

```python
df['outlooks']=outlooks
df
```

```python
temperatures=lbl_encode.fit_transform(df['temperature'])
temperatures
```

```python
df['temperatures']=temperatures
df
```

```python
humidities=lbl_encode.fit_transform(df['humidity'])
humidities
```

```python
df['humidities']=humidities
df
```

```python
windies=lbl_encode.fit_transform(df['windy'])
windies
```

```python
df['windies']=windies
df
```

```python
plays=lbl_encode.fit_transform(df['play'])
plays
```

```python
df['plays']=plays
df
```

```python
df.drop('outlook',axis= 1, inplace=True)
df.drop('humidity',axis=1, inplace=True)
df.drop('windy',axis=1, inplace=True)
df.drop('temperature',axis=1, inplace=True)
df.drop('play',axis=1, inplace=True)
```
