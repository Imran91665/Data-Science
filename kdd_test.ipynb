```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
```

```python
df=pd.read_csv(r'D:\data science\research practice\kddcup99 _1.csv')
```

```python
df.head(3)
```

```python
x=df.iloc[:,:39]
y=df['label']
```

```python
lbl_encode=LabelEncoder()
```

```python
y2=lbl_encode.fit_transform(y)
```

```python
protocol_types=lbl_encode.fit_transform(x['protocol_type'])
```

```python
x['protocol_types']=protocol_types
```

```python
x.drop('protocol_type',axis= 1, inplace=True)
```

```python
services=lbl_encode.fit_transform(x['service'])
```

```python
x['services']=services
```

```python
x.drop('service',axis= 1, inplace=True)
```

```python
flags=lbl_encode.fit_transform(x['flag'])
```

```python
x['flags']=flags
```

```python
x.drop('flag',axis= 1, inplace=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y2,test_size=0.30)
```

```python
model_1 = RandomForestClassifier(random_state=1)
```

```python
model_1.fit(x_train, y_train)
```

```python
model_1.score(x_test,y_test)
```

```python
pred=model_1.predict(x_test)
```

```python
accuracy_score(y_test,pred)
```

```python
print(classification_report(y_test,pred))
```
